## RECURRENT NETWORK

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
init_point = np.array([.3,.3]) ### CHANGED
weight_matrix = np.array([[-.8,.5],[-.1,.4]]) ### CHANGED
print(weight_matrix)
```

```python
times = range(1,10) ###
activation = init_point
activations = []
for t in times:  
    activation = activation.dot(weight_matrix)
    activations.append(activation)
points = np.asarray(activations)
```

```python
plt.plot(points[:,0], points[:,1])
plt.plot(points[:,0], points[:,1], 'bo')
for i, label in enumerate(times):
    plt.text(points[i,0], points[i,1], label)
```

```python
print(np.round(points,3))
```
